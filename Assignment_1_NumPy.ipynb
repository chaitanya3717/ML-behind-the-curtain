 Q6 Compute the inverse and primary diagnol sum of x=[[1,2],[3,4]]

```python
x=np.array([[1,2],[3,4]])
diag = np.diagonal(x)
print("the primary diagonal sum : ")
print(sum(diag))
inverse_x=np.linalg.inv(x)
print("inverse of matrix : ")
print(inverse_x)
print("diagonal sum of inverse: ")
print(sum(np.diagonal(inverse_x)))
```

Bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
Less than 3 –> ‘small’
3-5 –> ‘medium’
‘>=5 –> ‘large’

```python

bin = np.digitize(iris[:, 2].astype('float'), [0, 3, 5])


labels = {1: 'small', 2: 'medium', 3: 'large'}
text_array_petal_length = [labels[x] for x in bin]
print(text_array_petal_length)
```

Find the most frequent value of petal length (3rd column) in iris dataset

```python
l=list(map(float,iris[:,2]))
w= np.array(l)
uniqw, inverse = np.unique(w, return_inverse=True)
bin_count=np.bincount(inverse)
max_index_row = np.argmax(bin_count)

print(uniqw[max_index_row])
```
